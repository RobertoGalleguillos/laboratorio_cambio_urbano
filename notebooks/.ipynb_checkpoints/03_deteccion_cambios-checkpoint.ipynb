# 03_deteccion_cambios

Implementa m?todos de detecci?n de cambios entre fechas usando ?ndices espectrales calculados previamente.
Genera un raster clasificado `cambio_clasificado.tif` en `data/processed`.

```python
from pathlib import Path
import numpy as np
import rasterio
import matplotlib.pyplot as plt

repo_root = Path.cwd().parent
processed_dir = repo_root / 'data' / 'processed'
processed_dir.mkdir(parents=True, exist_ok=True)

t1_path = processed_dir / 'indices_2018.tif'
t2_path = processed_dir / 'indices_2024.tif'

if not t1_path.exists() or not t2_path.exists():
    raise FileNotFoundError('Faltan indices_2018.tif o indices_2024.tif en data/processed')
```

```python
def detectar_cambio_diferencia(ruta_t1: Path, ruta_t2: Path, umbral=0.15):
    """
    Detecta cambios usando diferencia de NDVI.
    Retorna: cambio (-1 perdida, 0 sin cambio, 1 ganancia), diferencia
    """
    with rasterio.open(ruta_t1) as src1:
        ndvi_t1 = src1.read(1)
        profile = src1.profile
    with rasterio.open(ruta_t2) as src2:
        ndvi_t2 = src2.read(1)

    diferencia = ndvi_t2 - ndvi_t1
    cambio = np.zeros_like(diferencia, dtype=np.int8)
    cambio[diferencia < -umbral] = -1
    cambio[diferencia > umbral] = 1

    pixeles_total = np.sum(~np.isnan(diferencia))
    perdida = np.sum(cambio == -1)
    ganancia = np.sum(cambio == 1)

    print(f'Pixeles con perdida: {perdida} ({100*perdida/pixeles_total:.1f}%)')
    print(f'Pixeles con ganancia: {ganancia} ({100*ganancia/pixeles_total:.1f}%)')
    print(f'Sin cambio significativo: {pixeles_total - perdida - ganancia}')

    return cambio, diferencia, profile
```

```python
cambio_diff, diff_ndvi, profile = detectar_cambio_diferencia(t1_path, t2_path, umbral=0.15)

# Guardar mapa de cambio (diferencia simple)
out_diff = processed_dir / 'cambio_diferencia_ndvi.tif'
profile.update(dtype='int8', count=1)
with rasterio.open(out_diff, 'w', **profile) as dst:
    dst.write(cambio_diff.astype('int8'), 1)
print('Guardado:', out_diff)
```

```python
def leer_indices(ruta: Path):
    with rasterio.open(ruta) as src:
        ndvi = src.read(1)
        ndbi = src.read(2)
        ndwi = src.read(3)
        profile = src.profile
    return {'ndvi': ndvi, 'ndbi': ndbi, 'ndwi': ndwi}, profile

def clasificar_cambio_urbano(indices_t1, indices_t2, umbrales=None):
    """
    Clases:
    0: Sin cambio
    1: Urbanizacion (vegetacion -> construido)
    2: Perdida de vegetacion
    3: Ganancia de vegetacion
    4: Nuevo cuerpo de agua
    5: Perdida de agua
    """
    if umbrales is None:
        umbrales = {
            'ndvi_veg': 0.3,
            'ndbi_urbano': 0.0,
            'cambio_min': 0.1,
        }

    ndvi_t1 = indices_t1['ndvi']
    ndbi_t1 = indices_t1['ndbi']
    ndwi_t1 = indices_t1['ndwi']
    ndvi_t2 = indices_t2['ndvi']
    ndbi_t2 = indices_t2['ndbi']
    ndwi_t2 = indices_t2['ndwi']

    clase = np.zeros_like(ndvi_t1, dtype=np.uint8)

    era_vegetacion = ndvi_t1 > umbrales['ndvi_veg']
    es_urbano = ndbi_t2 > umbrales['ndbi_urbano']
    clase[era_vegetacion & es_urbano] = 1

    perdio_veg = (ndvi_t1 - ndvi_t2) > umbrales['cambio_min']
    clase[(perdio_veg) & (clase == 0)] = 2

    gano_veg = (ndvi_t2 - ndvi_t1) > umbrales['cambio_min']
    clase[(gano_veg) & (clase == 0)] = 3

    era_no_agua = ndwi_t1 < 0
    es_agua = ndwi_t2 > 0.1
    clase[(era_no_agua) & (es_agua) & (clase == 0)] = 4

    era_agua = ndwi_t1 > 0.1
    no_es_agua = ndwi_t2 < 0
    clase[(era_agua) & (no_es_agua) & (clase == 0)] = 5

    return clase
```

```python
indices_t1, profile = leer_indices(t1_path)
indices_t2, _ = leer_indices(t2_path)

cambio_clase = clasificar_cambio_urbano(indices_t1, indices_t2)

# Guardar mapa clasificado
out_clase = processed_dir / 'cambio_clasificado.tif'
profile.update(dtype='uint8', count=1)
with rasterio.open(out_clase, 'w', **profile) as dst:
    dst.write(cambio_clase.astype('uint8'), 1)
print('Guardado:', out_clase)
```

## Visualizacion del cambio clasificado
Clases: 0 Sin cambio, 1 Urbanizacion, 2 Perdida veg., 3 Ganancia veg., 4 Nuevo agua, 5 Perdida agua.


```python
import matplotlib.colors as mcolors

cmap = mcolors.ListedColormap(['lightgray', 'orangered', 'gold', 'limegreen', 'deepskyblue', 'navy'])
labels = ['Sin cambio', 'Urbanizacion', 'Perdida veg.', 'Ganancia veg.', 'Nuevo agua', 'Perdida agua']

plt.figure(figsize=(6, 6))
plt.imshow(cambio_clase, cmap=cmap, vmin=0, vmax=5)
plt.title('Cambio clasificado (2018-2024)')
plt.axis('off')

handles = [plt.Line2D([0], [0], marker='s', color='w', label=labels[i],
                      markerfacecolor=cmap.colors[i], markersize=10)
           for i in range(len(labels))]
plt.legend(handles=handles, loc='lower right', frameon=True)
plt.show()
```

## Metodo 3 (opcional): Anomalias temporales
Usa todas las fechas disponibles para detectar anomalias en la ultima fecha.


```python
def detectar_anomalias_temporales(stack_indices, fecha_analisis=-1):
    historico = np.delete(stack_indices, fecha_analisis, axis=0)
    media_hist = np.nanmean(historico, axis=0)
    std_hist = np.nanstd(historico, axis=0)

    actual = stack_indices[fecha_analisis]
    z_score = (actual - media_hist) / (std_hist + 1e-10)

    cambio_significativo = np.abs(z_score) > 2
    direccion = np.zeros_like(z_score, dtype=np.int8)
    direccion[z_score < -2] = -1
    direccion[z_score > 2] = 1

    return z_score, cambio_significativo, direccion

# Stack NDVI de todas las fechas
ndvi_stack = []
for p in sorted(processed_dir.glob('indices_*.tif')):
    with rasterio.open(p) as src:
        ndvi_stack.append(src.read(1))

ndvi_stack = np.stack(ndvi_stack, axis=0)
z_score, cambio_sig, direccion = detectar_anomalias_temporales(ndvi_stack, fecha_analisis=-1)

plt.figure(figsize=(6, 6))
plt.imshow(z_score, cmap='coolwarm')
plt.title('Anomalia NDVI (z-score)')
plt.axis('off')
plt.show()
```
