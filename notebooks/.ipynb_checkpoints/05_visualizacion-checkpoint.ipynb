# 5. Visualización y Análisis Final de los Resultados

Este notebook tiene como objetivo consolidar y presentar de forma visual los principales resultados obtenidos en las etapas anteriores del laboratorio. A partir de los productos generados (índices espectrales, detección de cambios y análisis zonal), se elaboran mapas y gráficos finales que facilitan la interpretación del cambio urbano en el área de estudio. Estas visualizaciones están diseñadas para ser reutilizadas tanto en el informe final como en el dashboard interactivo.


## 5.1 Preparación del Entorno y Verificación de Insumos

En esta sección se configuran las rutas de trabajo y se cargan las librerías necesarias para la visualización de los resultados. Además, se verifica la existencia de los archivos generados en los notebooks anteriores, asegurando la correcta ejecución del flujo de trabajo y la portabilidad del notebook en distintos entornos computacionales.


```python
from pathlib import Path
import rasterio
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

repo_root = Path.cwd().parent
processed_dir = repo_root / "data" / "processed"
vector_dir = repo_root / "data" / "vector"
outputs_dir = repo_root / "outputs"
figures_dir = outputs_dir / "figures"
maps_dir = outputs_dir / "maps"

figures_dir.mkdir(parents=True, exist_ok=True)
maps_dir.mkdir(parents=True, exist_ok=True)

required = [
    processed_dir / "cambio_clasificado.tif",
    processed_dir / "estadisticas_indices.csv",
    vector_dir / "manzanas_censales.shp"
]

for f in required:
    if not f.exists():
        raise FileNotFoundError(f"Falta archivo requerido: {f}")

print("✔ Entorno listo para visualización")
```

## 5.2 Visualización del Mapa de Cambio Urbano Clasificado

Esta sección presenta el mapa final de cambio urbano clasificado, resultado del proceso de detección de cambios multitemporal. El mapa permite identificar espacialmente las principales dinámicas territoriales, tales como urbanización, pérdida o ganancia de vegetación y cambios en cuerpos de agua, proporcionando una visión integral del fenómeno estudiado.


```python
cambio_path = processed_dir / "cambio_clasificado.tif"

with rasterio.open(cambio_path) as src:
    cambio = src.read(1)

cmap = mcolors.ListedColormap([
    "lightgray", "orangered", "gold",
    "limegreen", "deepskyblue", "navy"
])

labels = [
    "Sin cambio", "Urbanización", "Pérdida vegetación",
    "Ganancia vegetación", "Nuevo cuerpo de agua", "Pérdida de agua"
]

plt.figure(figsize=(8, 8))
plt.imshow(cambio, cmap=cmap, vmin=0, vmax=5)
plt.title("Mapa de cambio urbano clasificado (2018–2024)")
plt.axis("off")

handles = [
    plt.Line2D([0], [0], marker='s', linestyle='',
               markerfacecolor=cmap.colors[i],
               label=labels[i], markersize=10)
    for i in range(len(labels))
]

plt.legend(handles=handles, loc="lower right")
plt.savefig(maps_dir / "cambio_clasificado.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 5.3 Evolución Temporal de Índices Espectrales

En esta sección se analizan las variaciones temporales de los principales índices espectrales (NDVI y NDBI) a lo largo del período de estudio. Los gráficos generados permiten observar tendencias asociadas a la expansión urbana y a los cambios en la cobertura vegetal, complementando el análisis espacial con una perspectiva temporal.


```python
df_temp = pd.read_csv(processed_dir / "estadisticas_indices.csv", index_col=0)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(df_temp.index, df_temp["ndvi_mean"], marker="o")
ax[0].set_title("Evolución NDVI medio")
ax[0].set_ylabel("NDVI")

ax[1].plot(df_temp.index, df_temp["ndbi_mean"], marker="s")
ax[1].set_title("Evolución NDBI medio")
ax[1].set_ylabel("NDBI")

for a in ax:
    a.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(figures_dir / "indices_temporales.png", dpi=150)
plt.show()
```

## 5.4 Análisis Espacial de Zonas con Mayor Urbanización

Esta sección muestra la distribución espacial de las zonas que presentan mayores niveles de urbanización acumulada. A partir de los resultados del análisis zonal, se identifican y visualizan las áreas más impactadas, lo que permite priorizar sectores y apoyar la toma de decisiones en el ámbito de la planificación territorial.


```python
# Cargar datos zonales y estadisticas
zonas = gpd.read_file(vector_dir / "manzanas_censales.shp")
stats = pd.read_csv(processed_dir / "estadisticas_cambio.csv")

# Normalizar tipos de identificadores para evitar errores de merge
zonas["MANZENT"] = zonas["MANZENT"].astype(str)
stats["zona"] = stats["zona"].astype(str)

# Unir geometria con estadisticas zonales
zonas_stats = zonas.merge(
    stats,
    left_on="MANZENT",
    right_on="zona"
)

# Eliminar columna duplicada
zonas_stats = zonas_stats.drop(columns=["zona"])

# Seleccionar Top 10 zonas con mayor urbanizacion
top_urb = zonas_stats.nlargest(10, "urbanizacion_ha")

# Visualizacion
ax = top_urb.plot(
    column="urbanizacion_ha",
    cmap="Reds",
    legend=True,
    figsize=(8, 8),
    edgecolor="black"
)

ax.set_title("Top 10 zonas con mayor urbanización acumulada (ha)")
ax.axis("off")

plt.savefig(maps_dir / "top10_urbanizacion.png", dpi=150, bbox_inches="tight")
plt.show()

```

## 5.5 Síntesis e Interpretación de los Resultados

Las visualizaciones generadas en este notebook permiten identificar patrones claros de cambio urbano en el área de estudio durante el período analizado. En particular, el mapa de cambio clasificado evidencia una expansión significativa de áreas urbanizadas, concentrada en zonas específicas, lo que sugiere procesos de crecimiento urbano no homogéneos en el territorio.

La evolución temporal de los índices espectrales refuerza estas observaciones, mostrando una tendencia general al aumento del NDBI y, en ciertos sectores, una disminución del NDVI. Estos comportamientos son consistentes con procesos de urbanización progresiva y posibles transformaciones en la cobertura vegetal asociadas al desarrollo urbano.

Por otro lado, el análisis zonal permite identificar las zonas con mayor impacto acumulado, facilitando la priorización territorial y el análisis comparativo entre sectores. En conjunto, estos resultados proporcionan una base sólida para la construcción del dashboard interactivo, el cual permitirá explorar los cambios de manera dinámica, así como para la elaboración de conclusiones.
