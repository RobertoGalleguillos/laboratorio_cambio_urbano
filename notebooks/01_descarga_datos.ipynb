# 01_descarga_datos

Este notebook verifica si los archivos requeridos existen en `data/raw` y `data/vector`.
Si falta alguno, descarga todo desde Google Drive usando `scripts/download_sentinel.py --drive`.

```python
from pathlib import Path
import subprocess
import sys

repo_root = Path.cwd().parent
raw_dir = repo_root / 'data' / 'raw'
vector_dir = repo_root / 'data' / 'vector'

required_raw = [
    raw_dir / 'sentinel2_2018.tif',
    raw_dir / 'sentinel2_2020.tif',
    raw_dir / 'sentinel2_2022.tif',
    raw_dir / 'sentinel2_2024.tif',
]

required_vector = [
    vector_dir / 'limite_comuna.gpkg',
    vector_dir / 'red_vial.gpkg',
    vector_dir / 'manzanas_censales.cpg',
    vector_dir / 'manzanas_censales.dbf',
    vector_dir / 'manzanas_censales.prj',
    vector_dir / 'manzanas_censales.shp',
    vector_dir / 'manzanas_censales.shx',
    vector_dir / 'manzanas_censales.qmd',
]

required = required_raw + required_vector
missing = [p for p in required if not p.exists()]

if missing:
    print('Faltan archivos. Descargando desde Drive...')
    cmd = [sys.executable, str(repo_root / 'scripts' / 'download_sentinel.py'), '--drive']
    subprocess.check_call(cmd)
    missing = [p for p in required if not p.exists()]

if missing:
    print('Aun faltan archivos:')
    for p in missing:
        print('-', p)
else:
    print('Todo OK. Archivos presentes en sus carpetas.')
```
